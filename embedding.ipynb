```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
#define vocabulary size
voc_size=10000
```

```python
##one hot representation

# one_hot() is used ONLY to convert words â†’ integer indices.
# Embedding layer cannot work on strings.It requires integer indices.
one_hot_repr=[one_hot(words,voc_size)for words in sent]
one_hot_repr
```

```python
##word embedding representation
from tensorflow.keras.layers import Embedding 
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
#pad_sequence is require to convert all input sentences of the same length

sent_len=8
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_len)
print(embedded_docs)
```

```python
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_len))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
